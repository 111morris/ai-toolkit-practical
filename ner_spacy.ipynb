```python
# task 3: NER+sentiment with spaCy
!pip install spacy
import spacy

nlp = spacy.load("en_core_web_sm")

sample_reviews = [
    "The Sony headphones are excellent, great sound!",
    "This Samsung TV is terrible and broke after a week.",
    "Apple iPhone delivers as promised.",
    "Unknown brand, not good quality."
]

for review in sample_reviews:
    doc = nlp(review)
    print("\nReview:", review)
    # NER
    print("Entities:", [(ent.text, ent.label_) for ent in doc.ents])
    # simple sentiment
    review_lower = review.lower()
    if "excellent" in review_lower or "great" in review_lower or "good" in review_lower:
        sentiment = "Positive"
    elif "terrible" in review_lower or "broke" in review_lower or "bad" in review_lower:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print("Sentiment:", sentiment)
```
